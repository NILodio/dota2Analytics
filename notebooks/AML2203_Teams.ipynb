```python
# import only required libraries
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures



df_team = pd.read_csv('../data/raw/matches_20000.csv')
```

```python
df_team
```

```python
# Get Unique Ids From Randia and dire team


df_radian = df_team[['radiant_team_id', 'radiant_name']]
df_dire = df_team[['dire_team_id', 'dire_name']]

# Get Unique Ids From Randia and dire team

df_radian = df_radian.drop_duplicates()
df_dire = df_dire.drop_duplicates()

# Get Unique Ids From Randia and dire team

#change name columns

df_radian.columns = ['team_id', 'team_name']

df_dire.columns = ['team_id', 'team_name']


# Merge the two dataframes

df_teams = pd.concat([df_radian, df_dire])

# Drop duplicates

df_teams = df_teams.drop_duplicates()


## Save into Json with id and name

df_teams.to_json('../data/processed/teams.json', orient='records')

```
